```python
import qiskit
from qiskit import QuantumCircuit, QuantumRegister, ClassicalRegister
from qiskit_aer import AerSimulator
from qiskit.transpiler import generate_preset_pass_manager
from qiskit.visualization import plot_histogram
from qiskit_ibm_runtime import Options, Session, SamplerV2 as Sampler
from qiskit_ibm_runtime import QiskitRuntimeService
service = QiskitRuntimeService(name="qgss-2025")
```

```python
qr = QuantumRegister(5,name ="q") # 5 qubits, 3 for encoding, two for detecting errors
cr0 = ClassicalRegister(1,name="c0") # 2 bits for storing the measurements
cr1 = ClassicalRegister(1,name="c1") # 2 bits for storing the measurements
cr = ClassicalRegister(3,name="cr")
qc = QuantumCircuit(qr,cr0,cr1,cr) # Declaring the QuantumCircuit

qc.cx(0,1) # Encoding with a CNOT with 
qc.cx(0,2)
qc.barrier() # Encoding finalized

# Error detection
qc.cx(0,3)
qc.cx(1,3)

qc.cx(1,4)
qc.cx(2,4)

qc.barrier()

qc.measure(3,0)
qc.measure(4,1)

qc.barrier()
# Now we try to correct the bit flip error if there is any
with qc.if_test((cr0,0)) and qc.if_test((cr1,0)):
    pass
with qc.if_test((cr0,0)) and qc.if_test((cr1,1)):
    qc.x(2)
with qc.if_test((cr0,1)) and qc.if_test((cr1,1)):
    qc.x(1)
with qc.if_test((cr0,1)) and qc.if_test((cr1,0)):
    qc.x(0)

qc.barrier()
qc.draw("mpl")
```

```python
# Now we are gonna do a simulation in a noisy backend
real_backends = [
    #service.backend("ibm_brisbane"),
    #service.backend("ibm_sherbrooke"),
    service.backend("ibm_torino"),
]
# The code is generalized for the three main QPUs available for free use, nevertheless ibm_torino is the only one
# that supports if_test for AerSimulator
noisy_fake_backends = []
for backend in real_backends:
    noisy_fake_backends.append(AerSimulator.from_backend(backend))
print(f"The noisy simulators are {noisy_fake_backends}")
```

```python
from qiskit.circuit import IfElseOp

for noisy in noisy_fake_backends:
    backend.target.add_instruction(IfElseOp, name="if_else")
    backend = noisy
    pm = generate_preset_pass_manager(backend=backend, optimization_level=3)
    qc_isa = pm.run(qc)

    sampler = Sampler(mode=backend)
    sampler.options.experimental = {"execution_path" : "gen3-experimental"}
    count = sampler.run([qc_isa], shots = 1000).result()
```

```python
[count[0].data.c0.get_counts(), count[0].data.c1.get_counts()]
count
```

```python
classical_measurements = [count[0].data.c0.get_counts(), count[0].data.c1.get_counts()]
display(plot_histogram(classical_measurements,
                       legend=["c0","c1"],
                       title='Classical Bits Measurements'))
# Due to the size of the circuit retrieving the measurements of the qubits was not possible
```
